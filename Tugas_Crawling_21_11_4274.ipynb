```python
import os
import googleapiclient.discovery
import csv
import pandas as pd
```

```python
# Set API KEY from gooogle cloud APIs & Services
API_KEY = 'AIzaSyAMb6R5B8NEvwz-DX9EJHwbPQU0_cRyG44'

# Initialize the YouTube Data API client
youtube = googleapiclient.discovery.build('youtube', 'v3', developerKey=API_KEY)

# URL of the video you want to crawl this comments
video_url = 'https://www.youtube.com/watch?v=Xsfwy8GQgps'

# Extract the video ID from the URL
video_id = video_url.split('v=')[1]

# Create a CSV file to save the comments
csv_file = open('crawling.csv', 'w', newline='', encoding='utf-8')
csv_writer = csv.writer(csv_file)

# Write header row to the CSV file
csv_writer.writerow(['Author', 'Created At', 'Comment', 'Like Count'])

# Call the API to retrieve the video comments
next_page_token = None

while True:
    request = youtube.commentThreads().list(
        part='snippet',
        videoId=video_id,
        textFormat='plainText',
        maxResults=100,
        pageToken=next_page_token
    )

    response = request.execute()

    for item in response['items']:
        comment_data = item['snippet']['topLevelComment']['snippet']
        comment = comment_data['textDisplay']
        author = comment_data['authorDisplayName']
        created_at = comment_data['publishedAt']
        like_count = comment_data['likeCount']

        # Write comment data to the CSV file
        csv_writer.writerow([author, created_at, comment, like_count])

    next_page_token = response.get('nextPageToken')

    if not next_page_token:
        break

# Close the CSV file
csv_file.close()

print('Comments have been saved to crawling.csv')
```
